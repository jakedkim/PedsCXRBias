```python
import os 
predefined_seeds = [2358, 4563, 7894, 13289, 15892, 19756, 23890, 27456, 30890, 34905, 38764, 42136, 46578, 50349, 54892, 59871, 63912, 68354, 72109, 76845, 80492, 84910, 89357, 93450, 97910]
parent = '/home/jkim/research/peds_cxr/'

for seed in predefined_seeds:
    os.makedirs(parent + f'no_finding/metadata/aggregate_sex_old_nf/aggregate_sex_old_nf_seed_{seed}', exist_ok=True)
    os.makedirs(parent + f'no_finding/metadata/aggregate_sex_young_nf/aggregate_sex_young_nf_seed_{seed}', exist_ok=True)
    os.makedirs(parent + f'no_finding/metadata/aggregate_age_M_nf/aggregate_age_M_nf_seed_{seed}', exist_ok=True)
    os.makedirs(parent + f'no_finding/metadata/aggregate_age_F_nf/aggregate_age_F_nf_seed_{seed}', exist_ok=True)
    os.makedirs(parent + f'no_finding/experiment/metadata_count/seed{seed}', exist_ok=True)
    os.makedirs(parent + f'no_finding/experiment/metadata_count/seed{seed}', exist_ok=True)
```

```python
import pandas as pd

# Load the CSV file
for seed in predefined_seeds:
    for ratio in ['0', '25', '50', '75', '100']: 
        for sex in ['M', 'F']:
            file_path = parent + f'no_finding/metadata/aggregate_age_nf/aggregate_age_nf_seed_{seed}/age_{ratio}_seed_{seed}.csv'  # Update with the correct path
            df = pd.read_csv(file_path)

            # Keep only records where 'Patient Gender' is 'M'
            df_filtered = df[df['Patient Gender'] == sex]

            # Save the filtered DataFrame to a new CSV file
            filtered_file_path = parent + f'no_finding/metadata/aggregate_age_{sex}_nf/aggregate_age_{sex}_nf_seed_{seed}/age_{sex}_{ratio}_seed_{seed}.csv'  # Update with the desired path
            df_filtered.to_csv(filtered_file_path, index=False)

            print(f"{filtered_file_path}")
```
